```python
import pandas as pd
```

```python
venues = pd.read_csv("linked_mag_venues.csv")
venues.head()
```

```python
venues['score'] = 2 / (1 / venues['jw'] + 1 / venues['qgram'])
venues['score_rank'] = venues.groupby(['scholar_name', 'scholar_cat'])['score'].rank(ascending=False)
venues.head(20)
selected_venues = venues.query("score_rank == 1.0 and score > 0.9")
selected_venues.head()
```

```python
selected_venues.query("score < 1")
```

```python
pd.set_option('display.max_rows', 110)
selected_venues.groupby("scholar_cat").agg('count')
```

```python
selected_venues[['scholar_cat', 'scholar_name', 'mag_name']].to_gbq("field_model_replication.top_mag_venues", project_id='gcp-cset-projects')
```
